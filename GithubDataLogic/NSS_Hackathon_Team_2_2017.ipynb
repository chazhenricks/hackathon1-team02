```python
class GithubAccts(object):
    def __init__(self):
        super(GithubAccts, self).__init__()    
    
    @staticmethod
    def get_bs4_html(url):
        page = requests.get(url)
        page_html = {}
        if page.status_code == 200:
            page_html = BeautifulSoup(page.content, 'html.parser')

        return page_html
    
    @staticmethod
    def find_top_repos(page_html, username):
        top_repos = list()        
        
        for tag in page_html.find_all('div'):
            if 'pinned-repo-item-content' in tag.get('class', []):
                top_repos.append("https://github.com{}".format(tag.a['href']))

        if len(top_repos) == 0:
            for tag in page_html.find_all('span', {'class' : 'pinned-repo-item-content'}):
                top_repos.append("https://github.com{}".format(tag.a['href']))
                                        
        print("Here is {}'s top repos: {}".format(username, top_repos))
        return top_repos
    
    @staticmethod
    def get_users_contributions_in_previous_year(page_html, username):
        contributions = ''
        try:            
            for tag in page_html.find_all('div'):
                if 'js-contribution-graph' in tag.get('class', []):
                    for text in tag.find_all('h2')[0].text.split(' '):
                        contributions_num_arr = text.split(',')
                        actual_contributions_num = ''.join(contributions_num_arr)
                        if len(actual_contributions_num) > 0 and actual_contributions_num.isdigit():
                            print("Here is {}'s contributions in the past year: {}".format(username, actual_contributions_num))
                            contributions = actual_contributions_num
                            
        except AttributeError:
            return contributions

        return contributions
```

```python
github_accts = {}
```

```python
for username in usernames:
    github_accts[username] = {}

    html_url = user_html_url.format(username)
    user_html = GithubAccts.get_bs4_html(html_url)
    user_contributions = GithubAccts.get_users_contributions_in_previous_year(user_html, username)
    github_accts[username]['contributions'] = user_contributions
    time.sleep(1)
```

```python
for username in usernames:
    user_profile = github_accts.get(username, None)
    
    if user_profile is not None:
        html_url = user_html_url.format(username)
        user_html = GithubAccts.get_bs4_html(html_url)
        top_repos = GithubAccts.find_top_repos(user_html, username)
        github_accts[username]['top_repos'] = top_repos
        time.sleep(1)
```
